```python
import sympy as sp
```

```python
m1, m2, m3, M = sp.symbols('m1 m2 m3 M')
t = sp.symbols('t')
theta_m2 = sp.Function('theta_m2')(t)
theta_m3 = sp.Function('theta_m3')(t)
y_m2 = sp.Function('y_m2')(t)
y_M = sp.Function('y_M')(t)
theta_M = sp.symbols('theta_M')
l1, l2, l3 = sp.symbols('l1 l2 l3')
r = sp.symbols('r')
k1, k2, k3, k4, k5 = sp.symbols('k1 k2 k3 k4 k5')
c4, c5 = sp.symbols('c4 c5')
g = sp.symbols('g')
```

```python
x_m2 = (y_M - y_m2) * sp.tan(theta_M)
delta_l2 = (y_M - y_m2) / sp.tan(theta_M)
l3_new = sp.sqrt((l3 * sp.sin(theta_m3) - x_m2)**2 + (l3 * sp.cos(theta_m3) - (y_m2 - l1 * theta_m2 + y_M))**2)
delta_l3 = l3_new - l3
```

```python
ans = 0.5 * (m1 + m3) * (l1 * sp.diff(theta_m2, t))**2 + 0.5 * m2 * (sp.diff(y_m2, t))**2 + 0.5 * (m1 + m2 + m3) * (sp.diff(((y_M - y_m2) * sp.tan(theta_M)), t))**2 + 0.5 * M * (sp.diff(y_M, t))**2
T = ans
sp.pprint(T)
```

```python
ans = 0.5 * c4 * (sp.diff(delta_l3, t))**2 + 0.5 * c5 * (sp.diff(delta_l2, t))**2
D = ans
sp.pprint(D)
```

```python
ans = 0.5 * k1 * (y_m2 + l1 * sp.sin(theta_m2))**2 + \
      0.5 * k2 * (y_m2)**2 + \
      0.5 * k3 * (y_m2 - l1 * sp.sin(theta_m2))**2 + \
      0.5 * k4 * (delta_l3)**2 + \
      0.5 * k5 * (delta_l2)**2 + \
      m1 * g * (r + y_m2 + l1 * sp.sin(theta_m2)) + \
      m2 * g * (r + y_m2) + \
      m3 * g * (r + y_m2 - l1 * sp.sin(theta_m2)) + \
      M * g * (r + y_m2 + (delta_l2 + l2) * sp.cos(theta_M))
V = ans
sp.pprint(V)
```

```python
T_diff_theta_m2_dot = sp.diff(T, sp.diff(theta_m2, t))
T_diff_theta_m3_dot = sp.diff(T, sp.diff(theta_m3, t))
T_diff_y_m2_dot = sp.diff(T, sp.diff(y_m2, t))
T_diff_y_M_dot = sp.diff(T, sp.diff(y_M, t))
diff_T_diff_theta_m2_dot = sp.diff(T_diff_theta_m2_dot, t)
diff_T_diff_theta_m3_dot = sp.diff(T_diff_theta_m3_dot, t)
diff_T_diff_y_m2_dot = sp.diff(T_diff_y_m2_dot, t)
diff_T_diff_y_M_dot = sp.diff(T_diff_y_M_dot, t)
```

```python
T_diff_theta_m2 = sp.diff(T, theta_m2)
T_diff_theta_m3 = sp.diff(T, theta_m3)
T_diff_y_m2 = sp.diff(T, y_m2)
T_diff_y_M= sp.diff(T, y_M)
```

```python
D_diff_theta_m2_dot = sp.diff(D, sp.diff(theta_m2, t))
D_diff_theta_m3_dot = sp.diff(D, sp.diff(theta_m3, t))
D_diff_y_m2_dot = sp.diff(D, sp.diff(y_m2, t))
D_diff_y_M_dot = sp.diff(D, sp.diff(y_M, t))
```

```python
V_diff_theta_m2 = sp.diff(V, theta_m2)
V_diff_theta_m3 = sp.diff(V, theta_m3)
V_diff_y_m2 = sp.diff(V, y_m2)
V_diff_y_M = sp.diff(V, y_M)
```

```python
theta_m2_func = diff_T_diff_theta_m2_dot - T_diff_theta_m2 + D_diff_theta_m2_dot + V_diff_theta_m2
sp.pprint(theta_m2_func)
```

```python
theta_m3_func = diff_T_diff_theta_m3_dot - T_diff_theta_m3 + D_diff_theta_m3_dot + V_diff_theta_m3
sp.pprint(theta_m3_func)
```

```python
y_m2_func = diff_T_diff_y_m2_dot - T_diff_y_m2 + D_diff_y_m2_dot + V_diff_y_m2
sp.pprint(y_m2_func)
```

```python
y_M_func = diff_T_diff_y_M_dot - T_diff_y_M + D_diff_y_M_dot + V_diff_y_M
sp.pprint(y_M_func)
```

```python
Q1 = sp.symbols('Q1')
Q2 = sp.symbols('Q2')
Q3 = sp.symbols('Q3')
Q4 = sp.symbols('Q4')
```

```python
function1 = theta_m2_func - Q1
solution1 = sp.solve(function1, sp.diff(sp.diff(theta_m2, t), t))
solution1_simple = solution1[0].simplify()
sp.pprint(solution1_simple)
```

```python
function2 = theta_m3_func - Q2
solution2 = sp.solve(function2, sp.diff(theta_m3, t))
solution2_simple = solution2[0].simplify()
sp.pprint(solution2_simple)
```

```python
function3 = y_m2_func - Q3
solution3 = sp.solve(function3, sp.diff(sp.diff(y_m2, t), t))
solution3_simple = solution3[0].simplify()
sp.pprint(solution3_simple)
```

```python
function4 = y_M_func - Q4
solution4 = sp.solve(function4, sp.diff(sp.diff(y_M, t), t))
solution4_simple = solution4[0].simplify()
sp.pprint(solution4_simple)
```

```python
theta_m2_dot = sp.symbols('theta_m2_dot')
theta_m3_dot = sp.symbols('theta_m3_dot')
y_m2_dot = sp.symbols('y_m2_dot')
y_M_dot = sp.symbols('y_M_dot')
theta_m2_ddot = sp.symbols('theta_m2_ddot')
theta_m3_ddot = sp.symbols('theta_m3_ddot')
y_m2_ddot = sp.symbols('y_m2_ddot')
y_M_ddot = sp.symbols('y_M_ddot')
```

```python
replacement = {
    sp.Derivative(sp.Derivative(theta_m2, t), t): theta_m2_ddot,
    sp.Derivative(sp.Derivative(theta_m3, t), t): theta_m3_ddot,
    sp.Derivative(sp.Derivative(y_m2, t), t): y_m2_ddot,
    sp.Derivative(sp.Derivative(y_M, t), t): y_M_ddot,
    sp.Derivative(theta_m2, t): theta_m2_dot,
    sp.Derivative(theta_m3, t): theta_m3_dot,
    sp.Derivative(y_m2, t): y_m2_dot,
    sp.Derivative(y_M, t): y_M_dot,
    theta_m2: sp.symbols('theta_m2'),
    theta_m3: sp.symbols('theta_m3'),
    y_m2: sp.symbols('y_m2'),
    y_M: sp.symbols('y_M')
}
```

```python
solution1_simple = solution1_simple.subs(replacement)
solution2_simple = solution2_simple.subs(replacement)
solution3_simple = solution3_simple.subs(replacement)
solution4_simple = solution4_simple.subs(replacement)
```
